```python
import numpy as np
from preprocessing import CustomLabelEncoder
```

```python

# Créer une instance de CustomLabelEncoder
encoder = CustomLabelEncoder()

# Tester la méthode `fit`
print("## Tester la méthode `fit`")
# Exemple de données
data = np.array(['chat', 'chien', 'oiseau', 'chat', 'chien', 'chat'])

# Appliquer la méthode fit
encoder.fit(data)

# Afficher les mappings
print("Mapping:", encoder.mapping)
print("Inverse Mapping:", encoder.inverse_mapping)

# Tester la méthode `transform`
print("## Tester la méthode `transform`")
# Transformer les données
transformed_data = encoder.transform(data)
print("Transformed Data:", transformed_data)

# Tester la méthode `inverse_transform`
print("## Tester la méthode `inverse_transform`")
# Inverser la transformation
inverse_transformed_data = encoder.inverse_transform(transformed_data)
print("Inverse Transformed Data:", inverse_transformed_data)

# Tester la méthode `save`
print("## Tester la méthode `save`")
# Sauvegarder les mappings dans un fichier
encoder.save('encoder_mapping.csv')
print("Mappings saved to 'encoder_mapping.csv'")

# Tester la méthode `load`
print("## Tester la méthode `load`")
# Créer une nouvelle instance et charger les mappings depuis le fichier
new_encoder = CustomLabelEncoder()
new_encoder.load('encoder_mapping.csv')

# Afficher les mappings chargés
print("Loaded Mapping:", new_encoder.mapping)
print("Loaded Inverse Mapping:", new_encoder.inverse_mapping)
```

```python
import pandas as pd
import numpy as np
from preprocessing import CustomStandardScaler
from stats import custom_mean, custom_std

# Exemple de DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
}
df = pd.DataFrame(data)

# Créer une instance de CustomStandardScaler
scaler = CustomStandardScaler()

# Tester la méthode `fit`
print("## Tester la méthode `fit`")
scaler.fit(df)
print("Data Dict after fit:", scaler.data_dict)

# Tester la méthode `transform`
print("## Tester la méthode `transform`")
transformed_df = scaler.transform(df.copy())
print("Transformed DataFrame:\n", transformed_df)

# Tester la méthode `fit_transform`
print("## Tester la méthode `fit_transform`")
fit_transformed_df = scaler.fit_transform(df.copy())
print("Fit Transformed DataFrame:\n", fit_transformed_df)

# Tester la méthode `save`
print("## Tester la méthode `save`")
scaler.save('scaler_params.csv')
print("Scaler parameters saved to 'scaler_params.csv'")

# Tester la méthode `load`
print("## Tester la méthode `load`")
new_scaler = CustomStandardScaler()
new_scaler.load('scaler_params.csv')
print("Loaded Data Dict:", new_scaler.data_dict)

# Vérifier que la transformation avec le nouveau scaler donne le même résultat
print("## Vérifier la transformation avec le nouveau scaler")
new_transformed_df = new_scaler.transform(df.copy())
print("New Transformed DataFrame:\n", new_transformed_df)
```
